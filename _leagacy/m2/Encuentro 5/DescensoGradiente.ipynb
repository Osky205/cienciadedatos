```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import plotly.graph_objs as go
```

```python
X,z=make_regression(100,2,noise=20)
x,y=X[:,0],X[:,1]
N=20
xv=np.linspace(x.min(),x.max(),N)
yv=np.linspace(y.min(),y.max(),N)
```

```python
modelo_rl = LinearRegression(fit_intercept=False)
modelo_rl.fit(X,z)
```

$$ \Large  y \approx  \hat{f}(x,y)  $$


$$ \large \hat{f}(x,y) = \beta_1 x + \beta_2 y   $$

$$\large \epsilon =  z_{real} -  z_{pred}    $$


$$\large \epsilon =  z_{real} -  \large \hat{f}(x,y)    $$

$$\large \epsilon =   z_{real} - \beta_1 x + \beta_2 y   $$

$$ \sum \large \epsilon^2 = \sum ( z_{real} - \beta_1 x + \beta_2 y )^2  $$



$$ \sum \large \epsilon^2 = \sum ( z_{real} - \beta_1 x + \beta_2 y )^2 = L(\beta_1 x , \beta_2)$$

# Grafiquemos esta L

```python
b1v=np.linspace(-100,100)
b2v=np.linspace(-100,100)

costo=L(b1v,b2v)
```

```python

def L(b1v,b2v):    
    costo=np.zeros([b1v.size,b2v.size])
    
    for i,b1 in enumerate(b1v):
        for j,b2 in enumerate(b2v):
            
            costo[i,j] = ((z - (b1*x + b2*y))**2).sum()
            
    return(costo)        
            
    
    
    
```

```python
data2=go.Surface(
        x=b1v,
        y=b2v,
        z=costo/costo.max())

fig = go.Figure(data=data2)
fig.update_traces( contours = {
        "z": {"show": True, "start": 0, "end": 0.2, "size": 0.005 , "project_z" : True},

    })

fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))

fig.show()
```

```python
data=go.Surface(
        x=b1v,
        y=b2v,
        z=costo)

layout = go.Layout(showlegend=False)
fig = go.Figure(data=data, layout=layout)
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
```

```python
def wireframe(xv,yv):
    line_marker = dict(color='#0066FF', width=1)
    lines=[]
    for i,xi in enumerate(xv):    
        lines.append(go.Scatter3d(x=np.ones(xv.size)*xi, y=yv, z=L(np.ones(xv.size)*xi,yv)[0], mode='lines', line=line_marker))

    for j,yj in enumerate(yv):    
        lines.append(go.Scatter3d(x=xv, y=np.ones(xv.size)*yj, z=L(xv,np.ones(yv.size)*yj)[:,0], mode='lines', line=line_marker))
    
    layout = go.Layout(showlegend=False)
    fig = go.Figure(data=lines, layout=layout)
    fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))

    return fig
```

```python
wireframe(b1v,b2v)
```
