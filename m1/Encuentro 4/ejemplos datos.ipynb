## Apis

```python
url2='https://apitransporte.buenosaires.gob.ar/colectivos/vehiclePositions?json=1&client_id=47ebe094c6c04a278c6ffabdaa226d61&client_secret=95fBe9DA0E444D658cE6143Eb90F7Bbb'
```

```python
import requests
bondis = requests.get(url2).json()['_entity']
```

```python
lats=list()
lngs=list()
for x in bondis:
    lats.append(x['_vehicle']['_position']['_latitude'])
    lngs.append(x['_vehicle']['_position']['_longitude'])
dict = {'lat': lats, 'long': lngs}  

df2 = pd.DataFrame(dict)    
```

```python
df2
```

```python
import plotly.express as px

df_bondis=df2[0:300]

fig = px.scatter_mapbox(df_bondis, lat="lat", lon="long", zoom=9, height=600)
fig.update_layout(mapbox_style="carto-positron")
fig.show()
```
